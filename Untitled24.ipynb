```python
# importing needed libraries

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils import class_weight
import os
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. SETUP PARAMETERS  ---
#  folder that DIRECTLY contains the 6 class subdirectories (Bird-drop, Clean, Dusty, etc.).
DATASET_DIR = '/content/drive/MyDrive/Solar_Panel_Dataset/Faulty_solar_panel'

# direcction path
ROOT_DIR = DATASET_DIR

# Model and Data Parameters
IMG_SIZE = (224, 224)
BATCH_SIZE = 32
NUM_CLASSES = 6

# Training Epochs (Total = Initial + Fine-Tuning)
INITIAL_EPOCHS = 15
FINE_TUNE_EPOCHS = 25
TOTAL_EPOCHS = INITIAL_EPOCHS + FINE_TUNE_EPOCHS

# Fine-Tuning Layer Control
FINE_TUNE_LAYERS = 40

print(f"Total Epochs: {TOTAL_EPOCHS}")

# --- 2. DATA PREPARATION (Using Data Augmentation and Single Directory Split) ---
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
    validation_split=0.2 # 20% of data used for validation
)

# Initialize key variables outside try block to prevent NameError
train_generator = None
validation_generator = None
class_weights_dict = {}
history = None

try:
    print(f"Attempting to read data and split from: {ROOT_DIR}")

    # Prepare Training Generator (80% of data)
    train_generator = train_datagen.flow_from_directory(
        ROOT_DIR,
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        subset='training',
        shuffle=True
    )

    # Prepare Validation Generator (20% of data)
    validation_generator = train_datagen.flow_from_directory(
        ROOT_DIR,
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        subset='validation',
        shuffle=False # Do not shuffle validation data
    )

    # --- 3. CALCULATE CLASS WEIGHTS (CRUCIAL FOR IMBALANCE FIX) ---
    labels = train_generator.classes
    unique_labels, counts = np.unique(labels, return_counts=True)

    weights = class_weight.compute_class_weight(
        class_weight='balanced',
        classes=unique_labels,
        y=labels
    )

    class_weights_dict = dict(zip(unique_labels, weights))

    print("\n--- Calculated Class Weights (Index: Weight) ---")
    # Print class names with weights
    class_names = list(train_generator.class_indices.keys())
    for index, weight in class_weights_dict.items():
        print(f"Class: {class_names[index]:<20} Weight: {weight:.2f}")
    print("--------------------------------------------------\n")

except Exception as e:
    print(f"\nFATAL ERROR: Data loading failed. Please check the following:")
    print(f"1. Correct Path: The path '{ROOT_DIR}' must exist in your Drive.")
    print(f"2. Class Folders: The 6 class folders MUST be directly inside '{ROOT_DIR}'.")
    print(f"The actual error encountered was: {e}")
    # Exit if data cannot be loaded to prevent further errors
    exit()


# --- 4. BUILD THE MOBILENETV2 TRANSFER LEARNING MODEL ---
base_model = MobileNetV2(
    input_shape=IMG_SIZE + (3,),
    include_top=False,
    weights='imagenet'
)
base_model.trainable = False

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.5)(x)
predictions = Dense(NUM_CLASSES, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=predictions)

base_learning_rate = 0.0001
model.compile(
    optimizer=Adam(learning_rate=base_learning_rate),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)


# --- 5. PHASE 1: INITIAL TRAINING (Feature Extraction) ---
print(f"\n--- PHASE 1: Training top layers for {INITIAL_EPOCHS} epochs ---")

history = model.fit(
    train_generator,
    epochs=INITIAL_EPOCHS,
    validation_data=validation_generator,
    class_weight=class_weights_dict,
    verbose=1
)


# --- 6. PHASE 2: FINE-TUNING THE ENTIRE MODEL ---
if history is not None:
    base_model.trainable = True

    # Fine-tuning the last 40 layers of MobileNetV2
    for layer in base_model.layers[:-FINE_TUNE_LAYERS]:
        layer.trainable = False

    fine_tune_learning_rate = base_learning_rate / 10
    model.compile(
        optimizer=Adam(learning_rate=fine_tune_learning_rate),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    print(f"\n--- PHASE 2: Fine-tuning last {FINE_TUNE_LAYERS} layers for {FINE_TUNE_EPOCHS} more epochs ---")

    history_fine = model.fit(
        train_generator,
        epochs=TOTAL_EPOCHS,
        initial_epoch=history.epoch[-1],
        validation_data=validation_generator,
        class_weight=class_weights_dict,
        verbose=1
    )

    # --- 7. MODEL EVALUATION (Using Validation Data) ---
    print("\n--- FINAL MODEL EVALUATION ON VALIDATION DATA ---")

    if validation_generator:
        # 7.1 Generate predictions on the hold-out validation set
        Y_pred = model.predict(validation_generator)
        y_pred = np.argmax(Y_pred, axis=1)
        y_true = validation_generator.classes

        class_labels = list(validation_generator.class_indices.keys())
        print("\n" + classification_report(y_true, y_pred, target_names=class_labels))

        cm = confusion_matrix(y_true, y_pred)
        plt.figure(figsize=(10, 8))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_labels, yticklabels=class_labels)
        plt.title('Confusion Matrix (Validation Data)')
        plt.ylabel('Actual Label')
        plt.xlabel('Predicted Label')
        plt.show()

    model_save_path = '/content/drive/MyDrive/solar_panel_defect_mobilenet.h5'
    try:
        model.save(model_save_path)
        print(f"\n✅ Model successfully saved to: {model_save_path}")
    except Exception as e:
        print(f"\nError saving model: {e}")
else:
    print("\nSkipping Phase 2 and Evaluation because Phase 1 training failed or was skipped.")
```
